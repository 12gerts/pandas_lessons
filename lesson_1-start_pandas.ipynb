1) Проверьте, сколько всего строк и столбцов имеется в датасете.

```python
taxi.shape
```

2) Давайте посмотрим на типы колонок. Все ли из них считались правильно? В качестве ответа выберите тип, преобладающий в датасете.

```python
taxi.dtypes.value_counts()
```

3) В названиях столбцов pcp 01, pcp 06, pcp 24 встречается пробел. Это не очень удобно, если в дальнейшем вы планируете обращаться к колонкам через точку, без использования кавычек и скобочек. 

Замените пробел в названиях на знак нижнего подчеркивания. 

```python
taxi = taxi.rename(columns=lambda c: c.replace('pcp ', 'pcp_'))
taxi
```

4) Cколько раз встречается каждый из районов?

```python
taxi['borough'].value_counts()
```

5) Посчитайте общее количество поездок.

```python
taxi['pickups'].sum()
```

6) А теперь сгруппируйте данные по району (borough) и укажите, из какого пункта было совершено больше всего поездок.

```python
taxi.groupby('borough').aggregate({'pickups' : 'sum'}).sort_values('pickups', ascending=False)
```

7) Возможно, в предыдущем степе вы просто посмотрели на данные или отсортировали значения. Ускорить данный процесс в дальнейшем помогут методы idxmin() и idxmax(), которые возвращают индекс минимального или максимального значения.

Сохраните название района с наименьшим числом поездок в переменную min_pickups, применив подходящий метод.

```python
min_pickups = taxi.groupby('borough').aggregate({'pickups' : 'sum'}).idxmin()
min_pickups
```

8) Сгруппируйте данные по двум признакам: району города и является ли день выходным (колонки borough и hday). Сравните среднее число поездок, и выберите районы, из которых по праздникам в среднем поступает больше заказов, чем в обычные дни. 

```python
taxi.groupby(['borough','hday'])['pickups'].mean()
```

9) Для каждого района посчитайте число поездок по месяцам. Отсортируйте полученные значения по убыванию.

```python
taxi.groupby(['borough', 'pickup_month'], as_index=False).aggregate({'pickups' : 'sum'}).sort_values('pickups', ascending=False)
```

## Минипроект

```python
# прочитаем датасет
hotel = pd.read_csv('2_bookings.csv', encoding='Windows-1251', sep=';')
hotel
```

1) Посмотрите на размеры таблицы. Сколько всего в ней столбцов?

```python
hotel.shape[1]
```

2) К какому типу/каким типам относится большинство переменных? 

```python
hotel.dtypes.value_counts()
```

3) Как вы уже могли заметить, при сборе датасета произошла небольшая ошибка, и названия колонок были записаны в неудобном для дальнейшей работы формате. Исправьте ситуацию, приведя названия столбцов к нижнему регистру и заменив пробелы на знак нижнего подчеркивания. 

```python
hotel = hotel.rename(columns=lambda c: c.lower().replace(' ', '_'))
hotel
```

4) Пользователи из каких стран совершили наибольшее число успешных бронирований? Бронирование считается успешным, если в дальнейшем не было отменено (переменная is_canceled). В качестве ответа выберите страны, входящие в топ-5.

```python
hotel.query('is_canceled == 0') \
    .groupby('country') \
    .agg({'is_canceled' : 'count'}) \
    .sort_values('is_canceled', ascending = False) \
    .head()
```

5)  На сколько ночей  в среднем бронируют отели типа City Hotel? Resort Hotel? Запишите полученные значения в пропуски с точностью до 2 знаков после точки.

```python
hotel.query('hotel == "City Hotel" or hotel == "Resort Hotel"') \
    .groupby('hotel') \
    .stays_total_nights \
    .mean() \
    .round(2)
```

6) Иногда тип номера, присвоенного клиенту (assigned_room_type), отличается от изначально забронированного (reserved_room_type). Такое может произойти, например, по причине овербукинга.

Сколько подобных наблюдений встретилось в датасете?

```python
hotel.query('reserved_room_type != assigned_room_type').shape[0]
```

7) На какой месяц чаще всего оформляли бронь в 2016 году? Изменился ли самый популярный месяц в 2017?

```python
hotel.groupby(['arrival_date_year', 'arrival_date_month']) \
    .aggregate({'hotel' : 'count'}) \
    .sort_values(by=['arrival_date_year','hotel'], ascending=[True, False])
```

8) Сгруппируйте данные по годам, а затем проверьте, на какой месяц бронирования отеля типа City Hotel отменялись чаще всего в 2015? 2016? 2017? 

```python
hotel.query('hotel == "City Hotel" and is_canceled == 1') \
    .groupby(['arrival_date_year', 'arrival_date_month'])['is_canceled'] \
    .count().to_frame() \
    .sort_values(by=['arrival_date_year', 'is_canceled'], ascending=False)
```

9) Посмотрите на числовые характеристики трёх колонок: adults, children и babies. Какая из них имеет наибольшее среднее значение?

```python
hotel[['adults', 'children', 'babies']].mean()
```

10) Создайте колонку total_kids, объединив столбцы children и babies. Для отелей какого типа среднее значение переменной оказалось наибольшим?

В качестве ответа укажите наибольшее среднее total_kids, округлив до 2 знаков после точки.

```python
hotel['total_kids'] = hotel['children'] + hotel['babies']
```

```python
hotel.groupby('hotel')['total_kids'].mean().round(2)
```

11) Не все бронирования завершились успешно (is_canceled), поэтому можно посчитать, сколько клиентов было потеряно в процессе. Иными словами, посчитать метрику под названием Churn Rate.

Churn rate (отток, коэффициент оттока) – это процент подписчиков (например, на push-уведомления от сайта), которые отписались от канала коммуникации, отказались от услуг сервиса в течение определенного периода времени. Иными словами, представляет собой отношение количества ушедших пользователей к общему количеству пользователей, выраженное в процентах.

В нашем случае Churn Rate - это процент клиентов, которые отменили бронирование. Давайте посмотрим, как эта метрика связана с наличием детей у клиентов!

Создайте переменную has_kids, которая принимает значение True, если клиент при бронировании указал хотя бы одного ребенка (total_kids), в противном случае – False. Далее проверьте, среди какой группы пользователей показатель оттока выше. 

```python
hotel['has_kids'] = (hotel['total_kids'] != 0) 
```

```python
(hotel.query('is_canceled == 1')[['has_kids', 'is_canceled']].value_counts() / hotel[['has_kids']].value_counts()*100).round(2)
```
