6) Приведите столбцы *start_at*, *end_at*, *arrived_at* к формату даты, используя `pd.to_datetime()`. Отфильтруйте данные и оставьте наблюдения со статусом заказа "asap" и "reserved" (*start_type*) .

```python
for column in ['start_at', 'end_at', 'arrived_at']:
    df[column] = pd.to_datetime(df[column])
```

```python
df[df["start_type"].isin(['asap', 'reserved'])]
```

7) Создайте колонку *wait_time*, в которой будет храниться разница между временем подачи машины (*arrived_at*) и временем заказа (*start_at*) в минутах.

```python
df['wait_time'] = (df['arrived_at'] - df['start_at']).astype('timedelta64[m]')
df['wait_time']
```

8) Посмотрим, какие водители опаздывали на заказы к определенному времени (*start_type* == 'reserved'). Опозданием считается время ожидания *wait_time* > 0.0.

Сгруппируйте отфильтрованные данные по *driver_id* и в качестве ответа укажите *id* водителя, который опоздал на заказ интересующего нас типа наибольшее число раз

```python
df.query('start_type == "reserved" and wait_time > 0.0').groupby('driver_id').agg({'user_id' : 'count'}).idxmax()
```

9) Добавьте столбец *month*, где будет храниться номер месяца, в котором был сделан заказ (*start_at*),  и столбец *weekday* – название дня недели. 

```python
df['month'] = df['start_at'].dt.month
df['weekday'] = df['start_at'].dt.strftime('%A')
```

## Минипроект



На этот раз данные имеют следующую структуру:

записываются для каждого пользователя, совершившего покупки, каждый день
для каждой даты есть своя папка, внутри неё – папки для каждого пользователя
внутри каждой папки есть файл data.csv, где и хранятся данные
Схематично выглядит так:

└── data  
...├── 2020-12-30  
...│  ├── FirstName_LastName1  
...│  │   └── data.csv  
...│  ├── FirstName_LastName2  
...│  │   └── data.csv  
...│  └── FirstName_LastName3  
...│      └── data.csv  
...└── 2020-12-31  
......├── FirstName_LastName1  
......│   └── data.csv  
......└── FirstName_LastName5  
..........└── data.csv  
          
Например, 30 декабря три покупателя сделали покупки, 31 – два 
(папки *2020-12-30* и *2020-12-31* соответственно). Поскольку клиент *FirstName_LastName1* купил товары в оба дня, для него имеется папка в папке для каждой из дат. Для других клиентов – по одной.

1) Соберите все данные из папки *data* в один датафрэйм, имеющий следующие столбцы: колонки из самих файлов (*product_id*, *quantity*), имя пользователя (*name*), и дату этих покупок (*date*), соответствует названию папки, где лежит папка с пользователем). В качестве ответа укажите сумму по колонке quantity

```python
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
os.listdir('data')
```

```python
dfs = []
for path, dirs, files in os.walk('data'):
    if len(files):
        arr = path.split('\\')
        dfs.append(pd.read_csv(f'{path}\\{files[0]}'))
        dfs[-1]['name'] = arr[2]
        dfs[-1]['date'] = arr[1]
               
df = pd.concat(dfs[:])
df
```

```python
df = df.reset_index(drop=True).drop(df.columns[[0]], axis = 1)
```

```python
df.quantity.sum()
```

2) Выясните, какой пользователь купил больше всего товаров. Если их несколько, то перечислите имена через запятую с пробелом и в алфавитном порядке.

```python
df.groupby('name') \
    .agg({'quantity' : 'sum'}) \
    .query('quantity == quantity.max()') \
    .index
```

3) Найдите топ-10 товаров по числу проданных единиц за всё время и постройте барплот (столбчатую диаграмму, `sns.barplot`), где:

по оси **x** – идентификаторы товаров (*product_id*)  

по оси **y** – суммарное число их покупок (сколько товаров с таким *product_id* было продано)  

Для практики попробуйте изменить параметры графика: цвет (`color`), прозрачность (`alpha`), сортировку значений по оси х (`order`). 

В качестве ответа укажите, сколько было продано единиц товара с *product_id* равным 56.

```python
top_df = df.groupby('product_id', as_index=False) \
    .agg({'quantity' : 'sum'}) \
    .nlargest(10, 'quantity') 
top_df
```

```python
plt.figure(figsize=(16,9))
sns.barplot(x=top_df.product_id, y=top_df.quantity, order=top_df.product_id, alpha=0.5, color='blue')
```

4) Посмотрим на продажи по дням! Для визуализации снова используйте барплот, только теперь по оси x будут дни, по которым у нас есть данные (*date*)

```python
day_df = df.groupby('date', as_index=False).agg({'quantity' : 'sum'})
day_df
```

```python
plt.figure(figsize=(16,9))
sns.barplot(x=day_df.date, y=day_df.quantity, order=day_df.date, alpha=0.5, color='salmon')
```

5) Сколько пользователей приобрели какой-либо товар повторно (более 1 раза)? Повтором будем считать покупку товара с одинаковым *product_id*, совершенную в разные дни. 

Например, пользователь Sasha Tsarev дважды приобрел товар 6. Поскольку покупка была совершена в один и тот же день, повторной в данном случае она считаться не будет. Katya Skriptsova также сделал две покупки товара 7, но в разные дни, поэтому этот случай засчитывается.

```python
df.drop_duplicates(subset=['product_id', 'name', 'date']) \
    .groupby(['name','product_id']) \
    .agg({'date' : 'count'}) \
    .query('date > 1') \
    .count()
```
